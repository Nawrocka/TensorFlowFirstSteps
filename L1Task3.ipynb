```
import tensorflow as tf

imdb = tf.keras.datasets.imdb

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(
    path='imdb.npz', num_words=None, skip_top=0, maxlen=None, seed=113,
    start_char=1, oov_char=2, index_from=3
)

print(f'Zbiór uczący: {x_train.shape}, zbiór walidacyjny: {x_test.shape}')
```

```
#Przygotowanie danych do modelu, padding:
maxlen = 100
x_train_pad = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen = maxlen)
x_test_pad = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen = maxlen)

x_train_pad.shape
```

```
x_train_pad[0]
```

```
#Sprawdzamy aktualną maksymalną liczbę zakodowanych słów, txn. kod słowa z najwyższą wartością oznacza, że tyle jest słów, bo zaczynamy od 1

max_features = max([max(x) for x in x_train_pad] + [max(x) for x in x_test_pad]) + 1
max_features
```

```
#Aby rozpocząć tworzenie modelu warto zobaczyć jakie mamy dane w y

y_train
```

```
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Embedding, Dense, LSTM, add

model = Sequential()
model.add(Embedding(max_features, 128)) #....
model.add(LSTM(64, dropout=0.2, reccurent_dropout=0.2))
model.add(Dense(1, activation='sigmoid')) #y_train są to wartości binarne, dlatego najlepiej przyjąć dla bin wartości funkcjęw kształcie "S"

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```
